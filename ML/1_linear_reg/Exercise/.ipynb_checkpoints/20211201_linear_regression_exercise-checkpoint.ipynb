```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
df.rename(columns={1: "income"})
df.head()
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(df.year, df['per capita income (US$)'], color='red', marker='+')
```

```python
new_df = df.drop('per capita income (US$)', axis='columns')
income = df['per capita income (US$)']
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df, income)
```

```python
new_years = [[2017],[2018],[2019],[2020], [2021]]
result = reg.predict(new_years)
result
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(df.year, df['per capita income (US$)'], color='red', marker='+')
plt.scatter(new_years, result, color='blue', marker='x')
```
